```
import numpy as np
import tensorflow as tf
import pandas as pd
```

```
tf.__version__
```

```
dataset=pd.read_csv('/content/drive/My Drive/Section 4 - Building an ANN/Churn_Modelling.csv')
x = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```
print(x)
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
print(x)
```

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
pd.set_option('max_colwidth',1000)
x
```

```
#np.savetxt('test.out', x, delimiter=',')   # X is an array
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```
from sklearn.preprocessing import  StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```
print(x_train)
```

#**Building the ANN**


### **Initializing the ANN**

```
ann = tf.keras.models.Sequential( )
```

### **Adding the input layer and the first hidden layer**

```
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
```

### **Adding the second hidden layer**

```
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
```

### **Adding the output layer**

```
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

# **Training the ANN**

### **Compiling the ANN**

```
ann.compile(optimizer='adam', loss='binary_crossentropy' , metrics=['accuracy'])
```

### **Training the ANN on the Training set**

```
ann.fit(x_train,y_train, batch_size = 64, epochs = 100)
_, accuracy = ann.evaluate(x_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

# **Making the predictions and evaluating the model**

### **Predicting the result of a single observation**

```
test = np.array([[1,0,0,600,1,40,3,60000,2,1,1,50000]])
print(test)
```

```
test = sc.transform(test)
```

```
output = ann.predict(test)
print(output)
```

```
print(output>0.5)
```

### **Predicting the Test set results**

```
y_pred = ann.predict(x_test)
y_pred = (y_pred>0.5)
```

### **Making the Confusion Matrix**

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
